```python
import pandas as pd
import streamlit as st
```

```python
hapvida = pd.read_csv(r'..\resources\RECLAMEAQUI_HAPVIDA.csv')
ibyte = pd.read_csv(r'..\resources\RECLAMEAQUI_IBYTE.csv')
nagem = pd.read_csv(r'..\resources\RECLAMEAQUI_NAGEM.csv')
```

```python
# Adicionar uma coluna para identificar a empresa
hapvida['empresa'] = 'Hapvida'
ibyte['empresa'] = 'Ibyte'
nagem['empresa'] = 'Nagem'
```

```python
# Combinar os datasets
df = pd.concat([hapvida, ibyte, nagem])
```

```python
empresa_selecionada = st.selectbox("Selecione a empresa", df['empresa'].unique())
estado_selecionado = st.selectbox("Selecione o estado", df['LOCAL'].str.split(' - ').str[-1].unique())
status_selecionado = st.selectbox("Selecione o status", df['STATUS'].unique())
tamanho_texto = st.slider("Selecione o tamanho do texto (descrição)", 0, df['DESCRICAO'].str.len().max())
```

```python
# Aplicar filtros no DataFrame
df_filtrado = df[
    (df['empresa'] == empresa_selecionada) &
    (df['LOCAL'].str.contains(estado_selecionado)) &
    (df['STATUS'] == status_selecionado) &
    (df['DESCRICAO'].str.len() <= tamanho_texto)
]
```

```python
# Converter ANO, MES e DIA em formato datetime
df_filtrado['DATA'] = pd.to_datetime(df[['ANO', 'MES', 'DIA']].astype(str).agg('-'.join, axis=1), format='%Y-%m-%d', errors='coerce')
```
